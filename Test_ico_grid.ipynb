# Create visualisations for icosahedral grid.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import gzip
import pickle
from icosahedron import Icosahedron, rand_rotation_icosahedron, rand_rotation_matrix, plot_voronoi, plot_voronoi_charts
# import torch
# import torch.utils.data as data_utils
```

```python
ico = Icosahedron(r=2,rad=1,c=np.array([0,0,0]))
charts = ico.get_charts_cut()
regions, vertices = ico.get_voronoi_regions_vertices()
```

```python
data = np.arange(charts[...,0].flatten().shape[0]).reshape(charts.shape[:-1]) % (charts.shape[1]*charts.shape[2])
```

```python
data[0,3,3] = -40 # dirty trick to have colormaps not start with white - delete later
plot_voronoi_charts(data, regions, vertices,elev=30)
plt.axis('off')
plt.savefig("Images/Icosahedron.jpg")
plt.savefig("Images/Icosahedron.png")
plt.savefig("Images/Icosahedron.pdf")
plt.show()
```

## Test the rotation function

```python
ico = Icosahedron(r=4,rad=1,c=np.array([0,0,0]))

charts = ico.get_charts_cut()
regions, vertices = ico.get_voronoi_regions_vertices()
data = np.arange(charts[...,0].flatten().shape[0]).reshape(charts.shape[:-1]) % (charts.shape[1]*charts.shape[2])

plot_voronoi_charts(data, regions, vertices,elev=30)
plt.show()
plt.close()

rot = rand_rotation_matrix()
regions, vertices = ico.get_voronoi_regions_vertices(rot=rot)
plot_voronoi_charts(data, regions, vertices, elev=30)
plt.show()
plt.close()


rot = rand_rotation_matrix()
regions, vertices = ico.get_voronoi_regions_vertices(rot=rot)
plot_voronoi_charts(data, regions, vertices, elev=30)
plt.show()
plt.close()

rot = rand_rotation_matrix()
regions, vertices = ico.get_voronoi_regions_vertices(rot=rot)
plot_voronoi_charts(data, regions, vertices, elev=30)
plt.show()
plt.close()
```
